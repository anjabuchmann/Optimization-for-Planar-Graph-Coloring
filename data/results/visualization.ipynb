### Visualization of Results

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_results(x, n_min, n_max, ants_tf=True, back_tf=True): 

    x2 = [i for i in range(10, 100, 10)] + [i for i in range(100, 201, 25)]

    results = pd.read_csv('./results_runtime.csv')

    ilp = results.loc[n_min:n_max, "ILP"]
    ants = results.loc[n_min:n_max, "Ant Colony"]
    backtracking = results.loc[n_min:min(n_max, 13), "Algorithm"]

    # plot
    fig, ax = plt.subplots()

    ax.grid(visible=True)

    ax.plot(x, ilp, 's-', linewidth=1.0, label='Integer Linear Programming (ILP)')
    if ants_tf: 
        ax.plot(x, ants, 'o-', color='steelblue', linewidth=1.0, label='Ant Colony Optimization (ACO)')
    if back_tf: 
        ax.plot(x2, backtracking, 'v-', color='skyblue', linewidth=1.0, label='Backtracking Algorithm (BT)')

    ax.legend(fontsize='small')
    ax.set_title('Average Runtime Comparison', fontsize='medium')
    ax.set_xlabel('n', fontsize='medium')
    ax.set_ylabel('Average Runtime [s]', fontsize='medium')
    fig.tight_layout()

    plt.show()

    return 
```

```python
x = [i for i in range(10, 101, 10)]

plot_results(x, 0, 9)
```

```python
x = [i for i in range(10, 100, 10)] + [i for i in range(100, 250, 25)] + [i for i in range(250, 1001, 50)] 

plot_results(x, 0, 30, ants_tf=False)
```

```python
x = [i for i in range(10, 100, 10)] + [i for i in range(100, 250, 25)] + [i for i in range(250, 1001, 50)] + [i for i in range(1500, 7001, 500)]

plot_results(x, 0, 42, ants_tf=False, back_tf=False)
```
